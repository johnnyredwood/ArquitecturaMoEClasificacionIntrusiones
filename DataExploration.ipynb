```python
#Procedemos a efectuar Oversampling con diferentes tecnicas mas analisis de artefactos
#Undesampling no lo aplico debido a que en investigaciones he visto que es mas para casos de desbalance extremo (ej 99% a 1%) y en mi caso la distribucion es 78% a 22%

from imblearn.over_sampling import ADASYN
from sklearn.neighbors import LocalOutlierFactor

def adasyn_con_lof_filtering(X_train, y_train, target_class_ratio='not majority', lof_contamination=0.05):

    print(f"Original dataset: {X_train.shape}")
    
    adasyn = ADASYN(sampling_strategy=target_class_ratio, random_state=42)
    X_resampled, y_resampled = adasyn.fit_resample(X_train, y_train)
    
    num_original = X_train.shape[0]
    X_synthetic = X_resampled[num_original:]
    y_synthetic = y_resampled[num_original:]
    
    print(f"Post-ADASYN shape: {X_resampled.shape} (Genero {len(X_synthetic)} samples)")
    
    lof = LocalOutlierFactor(n_neighbors=20, contamination=lof_contamination)
    outlier_labels = lof.fit_predict(X_synthetic)
    
    valid_synthetic_mask = (outlier_labels == 1)
    artifacts_removed = len(outlier_labels) - np.sum(valid_synthetic_mask)
    
    print(f"Artifact Analysis: LOF detectados y removidos {artifacts_removed} outliers")
    
    X_clean_synthetic = X_synthetic[valid_synthetic_mask]
    y_clean_synthetic = y_synthetic[valid_synthetic_mask]
    
    X_final = np.vstack((X_train, X_clean_synthetic))
    y_final = np.hstack((y_train, y_clean_synthetic))
    
    print(f"Final dataset: {X_final.shape}")
    
    return X_final, y_final

X_train_oversampled_1, y_train_oversampled_1= adasyn_con_lof_filtering(X_train_br_scaled,y_train)
```

```python
from imblearn.over_sampling import KMeansSMOTE
from sklearn.cluster import MiniBatchKMeans

def kmeans_smote_safe_generation(X_train, y_train, target_class_ratio=0.5, n_clusters=100):

    print(f"Original dataset shape: {X_train.shape}")
    
    # We use MiniBatchKMeans for computational efficiency on large tabular datasets like CICIDS
    kmeans_estimator = MiniBatchKMeans(n_clusters=n_clusters, random_state=42)
    
    # 1. Apply K-Means SMOTE
    # imbalance_ratio_threshold defines what is considered a "safe" cluster
    kmsmote = KMeansSMOTE(
        sampling_strategy=target_class_ratio, 
        random_state=42, 
        kmeans_estimator=kmeans_estimator,
        imbalance_ratio_threshold=1.0, # Adjust stricter (<1) to enforce higher minority purity
        k_neighbors=5
    )
    
    X_resampled, y_resampled = kmsmote.fit_resample(X_train, y_train)
    
    # 2. Artifact/Distribution Analysis
    num_original = X_train.shape[0]
    X_synthetic = X_resampled[num_original:]
    
    print(f"Resampled dataset shape: {X_resampled.shape} (Safely generated {len(X_synthetic)} samples)")
    
    # 3. Variance check: Ensure synthetic data didn't collapse into a single point (dead space)
    synth_variance = np.var(X_synthetic, axis=0)
    zero_variance_features = np.sum(synth_variance < 1e-6)
    
    print("\nArtifact Analysis: Variance Check")
    print(f"Features with near-zero variance in synthetic data: {zero_variance_features} out of {X_synthetic.shape[1]}")
    if zero_variance_features > 0:
        print("Warning: High number of zero-variance features suggests oversampling in overly tight clusters.")

    return X_resampled, y_resampled
```

```python
from sklearn.decomposition import PCA

def analisisPCA(xSet, tipoImputer, tipoFeatureSelection=None):

    print(f"ANÁLISIS PCA para datset con Imputer: {tipoImputer} y Feature Selection: {tipoFeatureSelection}")

    num_cols = xSet.select_dtypes(include="number").columns

    corr = pd.DataFrame(xSet, columns=num_cols).corr()

    plt.figure(figsize=(12,8))
    sns.heatmap(corr, cmap="coolwarm", center=0)
    plt.title("Matriz de correlación")
    plt.show()

    threshold = 0.8
    upper_triangle = corr.where(
        np.triu(np.ones(corr.shape), k=1).astype(bool)
    )

    strong_corr = upper_triangle.stack().abs()
    strong_corr = strong_corr[strong_corr > threshold]

    print(f"Número de correlaciones fuertes (>|{threshold}|): {len(strong_corr)}")

    eigenvalues, _ = np.linalg.eigh(corr)
    eigenvalues_sorted = sorted(eigenvalues, reverse=True)

    print("\nTop 10 eigenvalues:")
    print(eigenvalues_sorted[:10])

    pca = PCA()
    pca.fit(xSet)

    explained = pca.explained_variance_ratio_
    cumulative = np.cumsum(explained)

    # Número óptimo de componentes 95%
    n_components_95 = np.argmax(cumulative >= 0.95) + 1

    print("\nVarianza explicada por primeros 10 componentes:")
    print(explained[:10])

    print(f"\nComponentes necesarios para explicar 95% varianza: {n_components_95}")
    print(f"Reducción dimensional potencial: {X.shape[1]} reducido a {n_components_95}")

    plt.figure()
    plt.plot(cumulative)
    plt.axhline(y=0.95, linestyle='--')
    plt.axvline(x=n_components_95, linestyle='--')
    plt.xlabel("Número de componentes")
    plt.ylabel("Varianza acumulada")
    plt.title("Grafico - Varianza acumulada")
    plt.show()

    compression_ratio = n_components_95 / X.shape[1]
    print(f"\nRatio de compresión: {compression_ratio:.3f}")

    if compression_ratio < 0.6:
        print("PCA altamente justificado")
    elif compression_ratio < 0.8:
        print("PCA moderadamente justificado")
    else:
        print("PCA poco necesario")
```

```python
from sklearn.neighbors import NearestNeighbors
from scipy import stats
from imblearn.over_sampling import SMOTE

def detectar_artefactos_smote(X_train_original, y_train_original, X_train_smote, y_train_smote):
    #Funcion para detectar artefactos SMOTE

    # Separar clases originales y sintéticas
    clases_originales = np.unique(y_train_original)
    resultados = {}
    
    for clase in clases_originales:
        # Puntos originales de esta clase
        mask_original = (y_train_original == clase)
        X_original = X_train_original[mask_original]
        
        # Puntos SMOTE de esta clase con originales + sintéticos
        mask_smote = (y_train_smote == clase)
        X_smote = X_train_smote[mask_smote]

        if(len(X_original)==len(X_smote)):
            continue
        
        # Identificar puntos nuevos y comparar distancias
        nbrs = NearestNeighbors(n_neighbors=1).fit(X_original)
        distances, indices = nbrs.kneighbors(X_smote)
        
        # Puntos muy lejanos de cualquier punto original podrían ser artefactos
        umbral_distancia = np.percentile(distances, 95)  # percentil 95 como umbral
        puntos_sospechosos = distances.flatten() > umbral_distancia
        
        resultados[clase] = {
            'total_puntos_smote': len(X_smote),
            'puntos_originales': len(X_original),
            'puntos_sospechosos': puntos_sospechosos.sum(),
            'porcentaje_sospechosos': (puntos_sospechosos.sum() / len(X_smote)) * 100,
            'distancia_maxima': distances.max(),
            'distancia_media': distances.mean()
        }
    
    return pd.DataFrame(resultados).T
```

```python
#SMOTE con análisis de riesgo (artefactos)
#Pregunta para Ricardo aplico SMOTE en que clases minoritarias bajo que criterios
#Esta bien usar SMOTE que otras tecnicas de oversampling puedo usar sin generar contaminacion
#Acorde a investigaciones he visto SMOTE solo en clases medianas ya que SMOTE solo se aplica cuando hay suficiente densidad para interpolar

conteos = y_train.value_counts()

# clase mayoritaria
max_count = conteos.max()

# imbalance ratio
ir_por_clase = (max_count / conteos).sort_values()

df_ir = pd.DataFrame({"count": conteos, "IR": ir_por_clase}).sort_values("IR")

df_ir
```

```python
def clasificar_rareza(ir):

    if ir <= 10:
        return "balanceado"

    elif ir <= 100:
        return "minoria_smote_aceptable"

    else:
        return "clase_extrema_no_smote"
```

```python
df_ir["rareza"] = df_ir["IR"].apply(clasificar_rareza)

df_ir
```

```python
smote_labels = df_ir[df_ir["rareza"] == "minoria_smote_aceptable"].index.tolist()

no_smote_labels = df_ir[df_ir["rareza"] == "clase_extrema_no_smote"].index.tolist()

print("SMOTE candidatos:", smote_labels)
```

```python
#Estrategia automática para SMOTE

target_ir = 10
target_size = int(max_count / target_ir)

sampling_strategy = {}

for label in smote_labels:

    current = conteos[label]

    if current < target_size:
        sampling_strategy[label] = target_size

sampling_strategy
```

```python
# Obtengo data Pre SMOTE
X_train_pre_smote = X_train.copy()
y_train_pre_smote = y_train.copy()

# Aplicar SMOTE
smote = SMOTE(
    sampling_strategy=sampling_strategy,
    k_neighbors=5,
    random_state=42
)

X_train, y_train = smote.fit_resample(X_train, y_train)

# Analizar artefactos
analisis_artefactos = detectar_artefactos_smote(
    X_train_pre_smote, y_train_pre_smote, 
    X_train, y_train
)

print("Análisis de artefactos por clase:")
print(analisis_artefactos)
```
